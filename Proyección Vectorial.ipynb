¿Qué es una proyección Vectorial?
La proyección vectorial es la sombra que genera un vector sobre otro de tal manera que se forma un angulo recto entre el vextor y el vector de proyección, hay  proyección escalar que es la norma del vector de proyección

Y Se calcula de la siguiente manera:

Proy(u,v)= (u.v/v.v)*v



```python
import numpy as np
def proyec(u,v):
    prod_dot_uv=np.dot(u,v)
    norm_v=np.dot(v,v)
    proy_uv=((prod_dot_uv/norm_v)*v)
    return proy_uv
```

```python
proyec(np.array([2,4]),np.array([3,5]))
```
