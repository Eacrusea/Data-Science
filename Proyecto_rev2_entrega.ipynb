```python
def reg_lin(x1, y1, epochs, lr, m, b, print_every):
    error_list = []
    vector_m_b_list = []
    iter = 1
    vector = np.ones(len(x1))
    x1_trans = x1.T
    matriz = np.array([x1_trans, vector])
    matriz_trans = matriz.T
    
    while iter <= epochs:
        vector_m_b = np.array([m, b])
        y_hat = np.dot(matriz_trans, vector_m_b)
        
        diferencia_y = y1 - y_hat
        dif_cuadra = diferencia_y**2
        
        error = 0.5*np.sum(dif_cuadra)/len(dif_cuadra)
        
        gradiente_m = np.mean(diferencia_y*x1)
        gradiente_b = np.mean(diferencia_y)
        
        m_iter = m - lr*gradiente_m
        b_iter = b - lr*gradiente_b
        
        if iter % print_every == 0:
            print("Error: {}, m: {}, b: {}".format(error, m, b))
        
        m = m_iter
        b = b_iter
    
        error_list.append(error)
        vector_m_b_list.append(vector_m_b)
        
        iter += 1
        
    return np.array(error_list), np.array(vector_m_b_list)
        
```

```python
errors, m_b = reg_lin(x1, y, 16, .5, m1, b1, 3)
```

```python
epochs=16
m_b_opt=m_b[epochs-1]
m_b_opt
```

```python
def grafica_errores():
    curva_errores, m_b = reg_lin(x1, y, 16, .5, m1, b1, 3)
    equis=np.arange(1,len(curva_errores),1)
    grafica_e=plt.plot(equis,curva_errores,'o',label='Datos')
    return grafica_e
```

```python
def test(x1, y, m, b, generate_every):
    errors, m_b = reg_lin(x1, y, 16, .5, m, b, generate_every)
    
    for mb_index, current_mb in enumerate(m_b):
        if mb_index % generate_every == 0:
            result_f = []
            for x_index, current_x in enumerate(x1):    
                m = current_mb[0]
                b = current_mb[1]
                result_f.append(current_x * m + b)

            plt.plot(x1, y, '--', label='original', color='y')
            plt.plot(x1, result_f, 'o', label='datos', color='r')
            plt.show()
```
